```python

import requests
import pandas as pd
from IPython.display import display  # opsional, untuk notebook

def scan_ip_epep(ip=""):
    """
    Lacak IP dengan ip-api.com
    - ip kosong -> pakai IP publik sendiri
    - ip diisi -> lacak IP tersebut
    """
    ip = ip.strip()

    # Bangun endpoint yang benar
    if ip:
        url = f"http://ip-api.com/json/{ip}"
    else:
        url = "http://ip-api.com/json/"  # otomatis pakai IP requestor

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Accept": "application/json"
    }

    try:
        res = requests.get(url, headers=headers, timeout=10)
        res.raise_for_status()  # lempar error kalau status bukan 2xx

        # Pastikan response JSON
        if not res.headers.get("Content-Type", "").startswith("application/json"):
            print("❌ Response bukan JSON. Cek endpoint API.")
            return

        data = res.json()

        if data.get("status") == "fail":
            print(f"❌ Gagal lacak: {data.get('message', 'Unknown error')}")
            return

        # Siapkan data
        info = {
            "IP": data.get("query"),
            "Negara": data.get("country"),
            "Region": data.get("regionName"),
            "Kota": data.get("city"),
            "ISP": data.get("isp"),
            "Latitude": data.get("lat"),
            "Longitude": data.get("lon"),
            "Zona Waktu": data.get("timezone"),
            "Kode Pos": data.get("zip"),
        }
        # Filter yang None
        info = {k: v for k, v in info.items() if v is not None}

        print("\n🦅 [ INTEL TRACKING SELESAI - AUTO BOOYAH ] 🦅")
        df = pd.DataFrame(list(info.items()), columns=["Sektor", "Detail"])
        display(df)  # kalau di notebook

    except requests.exceptions.RequestException as e:
        print(f"❌ Gagal koneksi ke API: {e}")
    except ValueError:
        print("❌ Response bukan JSON valid.")
    except Exception as e:
        print(f"❌ Error tak terduga: {e}")

# Eksekusi
TARGET_IP = "121.187.15.63"  # @param {type:"string"}
scan_ip_epep(TARGET_IP)
```
